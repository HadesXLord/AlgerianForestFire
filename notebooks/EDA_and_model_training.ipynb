```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header=1)
df[df.isnull().any(axis=1)]
```

```python
df.loc[:122,'Region']=0
df.loc[122:,'Region']=1
df
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1)]
```

```python
df=df.drop(123)
df.iloc[121:124]
```

```python
df.info()
```

```python
df.columns=df.columns.str.strip()
df.columns
```

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']]=df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
df.info()
```

```python
for col in df.columns:
    if df[col].dtype=='object' and col!='Classes':
        df[col]=df[col].astype(float)

df.info()
```

```python
df['Classes']=np.where(df['Classes'].str.contains('not fire'),0,1)
```

```python
df
```

```python
df.info()
```

```python
df.drop(columns=['day','month','year'],inplace=True)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split

x=df.drop(columns=['FWI','Classes','Region'],axis=1)
y=df['FWI']
y_classifier=df['Classes']

x,y,y_classifier

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
x_train_classifier,x_test_classifier,y_train_classifier,y_test_classifier=train_test_split(x,y_classifier,test_size=0.2,random_state=42)
```

```python
x_train,y_train,x_test,y_test
```

Linear Regression with L2 Regularization(RidgeCV)

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import RidgeCV
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score


linear_model=make_pipeline(StandardScaler(),RidgeCV(cv=5))


linear_model.fit(x_train,y_train)

n=x_test.shape[0]
p=x_test.shape[1]

r2=r2_score(y_test,linear_model.predict(x_test))
adjusted_r2=1-(1-r2)*(n-1)/(n-p-1)

print(f'R2 Score: {r2}')
print(f'Adjusted R2 Score: {adjusted_r2}')
linear_model.predict(x_test)
```

Logistic Regression (Binary Classifier)

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,accuracy_score

logistic_model=make_pipeline(StandardScaler(),LogisticRegression())
logistic_model.fit(x_train_classifier,y_train_classifier)

confusion_matrix(y_test_classifier,logistic_model.predict(x_test_classifier))
accuracy_score=accuracy_score(y_test_classifier,logistic_model.predict(x_test_classifier))

print(f'Accuracy Score: {accuracy_score}')

```

```python
import joblib

joblib.dump(linear_model,'linear_model.pkl')
joblib.dump(logistic_model,'logistic_model.pkl')
```
